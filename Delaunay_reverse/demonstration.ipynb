# Runtime analisis 

```python
def run(n, a, b):
    import time

    points = np.array([np.random.uniform(a, b, size=(2,)) for _ in range(n)])

    start = time.time()
    vor = Voronoi(points) # implemented with Delaunay triangulation
    end = time.time()

    return end - start
```

```python
plt.close()

times = [[], []]

ns = [10, 50, 100, 200, 400, 500, 1000, 3000, 5000, 7000, 10000, 12000]

for i in ns:
    time1 = run(i, 0, 30)
    time2 = run(i, 0, 30)
    time3 = run(i, 0, 30)
    
    time = np.mean([time1, time2, time3])
    std = np.std([time1, time2, time3])
    
    times[0].append(time)
    times[1].append(std)

plt.errorbar(ns, times[0], yerr=times[1], marker='o', ms=4, ecolor='red')
plt.show()
```

```python
def run(n, a, b):
    import time
    
    k = np.int(np.sqrt(n))
    
    x = np.linspace(a, b, num=k)
    y = np.linspace(a, b, num=k)
    
    points = [(x_i, y_i) for x_i in x for y_i in y]
    #points = np.array([np.random.uniform(a, b, size=(2,)) for _ in range(n)])

    start = time.time()
    vor = Voronoi(points) # implemented with Delaunay triangulation
    end = time.time()

    return end - start
```

```python
plt.close()

times = [[], []]

ns = [10, 50, 100, 200, 400, 500, 1000, 3000, 5000, 7000, 10000, 12000]

for i in ns:
    time1 = run(i, 0, 30)
    time2 = run(i, 0, 30)
    time3 = run(i, 0, 30)
    
    time = np.mean([time1, time2, time3])
    std = np.std([time1, time2, time3])
    
    times[0].append(time)
    times[1].append(std)

plt.errorbar(ns, times[0], yerr=times[1], marker='o', ms=4, ecolor='red')
plt.show()
```
